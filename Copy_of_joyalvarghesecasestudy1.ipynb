```python
from scipy.stats import ttest_ind
```

```python
data1=data['Sales before']
data2=data['Sales after']
data1
```

```python
data1.mean()
```

```python
data2
```

```python
data2.mean()
```

```python
t_test1,p_value1=ttest_ind(data1,data2)
t_test1
```

```python
p_value1
```

```python
if p_value1<0.05:
  print('reject H0')
else:
  print('H0 accept')
```

```python
#As the observed p_value is less than 0.05 the null hypothesis is rejected and hence it  seems sales
```
